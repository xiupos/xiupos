SSH 接続に利用される秘密鍵・公開鍵について, もし公開鍵 `~/.ssh/id_ed25519.pub` を紛失してしまい, 手元に秘密鍵 `~/.ssh/id_ed25519` しかないときは, 以下で公開鍵を生成することができる.

```sh
ssh-keygen -y -f ~/.ssh/id_ed25519 > ~/.ssh/id_ed25519.pub
```

これと同じことを数学的に実行するのがこの記事の本題.
